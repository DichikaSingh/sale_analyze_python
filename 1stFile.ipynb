```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("D:\diwali sales dataset.csv", encoding='unicode_escape') #pass unicode_escape to aviod encoding error
```

```python
#like here i can analyz the 2 column has unfilled data (status and unnamed) so i have to remove this fileds
df.drop(['Status','unnamed1'], axis=1, inplace=True) #here i pass the axis=1 as parmater so axis means is all row of dataframe
```

```python
df.isnull().sum() #this stataement add all the null values of fileds individually,here you can clearly see 12 null values are present in amount filed.
```

```python
df.dropna(inplace= True) #dropna basically deteled null data records from dataframe 
#i place parameter inplace=true cause if we dont pass it , then automatically deleted records show gain in dataframe so for save we should pass the inplace 
```

```python
df.shape #again i chcek the total records of Dataframe, clearly shows its deleted 12 nulls values record from df
```

```python
#now change the datatype of coloumn
df['Amount'] = df['Amount'].astype('int')
```

```python
#Count female buyers and male buyers
ax = sns.countplot(x='Gender', data=df) 

sns.set(rc={'figure.figsize':(7,5)})
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
#Total amount generated by man and female
df.groupby(['Gender'], as_index= False) ['Amount'].sum().sort_values(by= 'Amount', ascending= False)
```

```python
#Total Amount Generated by Gender Category
sales_gender = df.groupby(['Gender'], as_index= False) ['Amount'].sum().sort_values(by= 'Amount', ascending= False)
sns.set(rc={'figure.figsize':(5,5)})
ax = sns.barplot(x='Gender', y = 'Amount', data= sales_gender)
```

```python
# here we analyize most of the buyers are female even purchasing power of female is gretaer than man
```

```python

Age_group_amount = df.groupby(['Age Group'], as_index= False) ['Amount'].sum().sort_values(by= 'Amount', ascending= False)
ax = sns.barplot(x= 'Age Group', y= 'Amount' ,data= Age_group_amount)
```

```python
ax= sns.countplot(x='Age Group', data= df, hue='Gender')
```

```python
#expend money by age group
```

```python
sales_zone = df.groupby(['Zone'], as_index= False) ['Amount'].sum().sort_values(by= 'Amount', ascending= False)
ax = sns.barplot(x='Zone', y = 'Amount', data= sales_zone )
```

```python
ax = sns.countplot(x = 'Zone', data= df, hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.columns
```

```python
# Top 10 State by amount of Orders

top5_sales_by_state=df.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10) #retrive top 10

sns.set(rc={'figure.figsize':(20,5)}) #it is uesd to rezise the plot

sns.barplot(data=top5_sales_by_state, x= 'State', y = 'Amount')

#here we can analyz top1 sales/amount generated from UP and also analyz top1, top5, and top10
```

```python
# Top 10 State by amount of Orders

orders_by_state = df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by= 'Orders', ascending= False).head(10)

sns.set(rc={'figure.figsize':(20,5)})

sns.barplot(data= orders_by_state, x='State', y='Orders' )

#here we can analyz top1 orders voloumn from UP and also analyz top1, top5, and top10
```

```python
#Maternal Status Count in df
ax=sns.countplot(x='Marital_Status', data=df, hue= 'Gender')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Marital status by amount
amount_by_maternal_status = df.groupby(['Marital_Status', 'Gender'], as_index= False) ['Amount'].sum().sort_values(by='Amount', ascending= False)
sns.set(rc={'figure.figsize':(5,5)})
ax=sns.barplot(data=amount_by_maternal_status, x= 'Marital_Status', y='Amount', hue='Gender')

#here we can analyz most of female buyers are not married and same as most of the male buyers are not married
```

```python
#here we count no of orders placed based on occupations
sns.set(rc={'figure.figsize':(20,4)})
ax= sns.countplot(x='Occupation', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#here we analyz total amount by occupation and seperate by gender category
total_amount_by_occupation= df.groupby(['Occupation','Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending= False )
sns.set(rc={'figure.figsize':(20,4)})
sns.barplot(x='Occupation', y='Amount', data=total_amount_by_occupation, hue='Gender')

#here we cann analyze  from IT sector we generated highest amount of sales where female are most buyers.
```

```python
#count of orders by product-category
sns.set(rc={'figure.figsize':(25,5)})
ax=sns.countplot(x='Product_Category',data=df)
for bar in ax.containers:
    ax.bar_label(bar)


# you can see the most orders voulmn by product_category is clothing & apparel
```

```python
#total amount generated by product category
total_amount_by_productCategory = df.groupby(['Product_Category'], as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.barplot(x='Product_Category', y='Amount', data=total_amount_by_productCategory)

#we can see most top selling product is Food category and then we can also analyz top10 amount by Product category


```
